```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
```

```python
digits = load_digits()
digits.data.shape
```

```python
tsne = TSNE(n_components=2)
xt = tsne.fit_transform(digits.data)
```

```python
from sklearn.neighbors import NearestNeighbors
nbrs = NearestNeighbors(n_neighbors=10, algorithm='ball_tree').fit(digits.data)
distances,indices = nbrs.kneighbors(digits.data)
```

```python
distances.shape
```

```python
import numpy as np
lis = np.reshape(distances, (distances.shape[0]*distances.shape[1]))
```

```python
import matplotlib.pyplot as plt
s = sorted(lis)
plt.plot(s)
```

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=27, min_samples=10).fit(digits.data)
y_cls = clustering.labels_
plt.scatter(xt[:,0], xt[:,1],c=y_cls)
```
